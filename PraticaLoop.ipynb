```python
lista=[]
while True:
    nota=int(input("Digite uma nota entre 10 e 20"))
    if nota>=10 and nota<=20:
        lista.append(nota)
    else:
        break
print(f'A média das notas inseridas é {sum(lista)/len(lista)}')
```

```python
N=int(input('Digite o número do qual você quer saber os divisores'))
print(f"Os divisores do número {N} são ")
for numero in range(1,N+1):
    if N%numero==0:
        print(numero)
```

```python
N=int(input('Digite o número do qual você quer saber a soma dos divisiores, com exceção dele mesmo'))
sum=0
print(f"A soma dos divisores de {N} com exceção dele mesmo é ")
for numero in range(1,N):
    if N%numero==0:
        sum+=numero
print(sum)
```

```python
sum=0
for numero in range(1,1001):
    if numero%3==0 or numero%5==0:
        sum+=numero
print(f'A soma dos números que sejam múltiplos de 3 ou 5 até 1000 é {sum}')
```

```python
N=int(input('Digite um número'))
aux=N+1
while aux%11!=0 and aux%13!=0 and aux%17!=0:
    aux+=1
print(f'O primeiro número após {N} que seja múltiplo de 11,13 ou 17 é {aux}')
```
